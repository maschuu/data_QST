```python
from matplotlib import pyplot as plt

import tools_plotting as tools
```

Here we plot figure D1 of the paper.

```python
noise_type="amp_damping"
# we start from this local amplitude damping noise strength in the 6-qubit system
# as we increase the number of qubits we decrease the local amplitude damping noise strength to keep the effetive depolarizing strength peff constant
strength_6qubit=0.004
noise_position="after_layer"
samples=128
layers_max=40
# these are the numbers of qubits in the graphs
ns=[6,8,10,12]
# this is the degree of the graph, here we consider 3-regular graphs
d=3
# this is the probability of an edge creation in the Erdos Renyi graph
p_edge=0.5


tools.plot_pur_3reg_renyi(layers_max=layers_max,ns=ns,d=d,p_edge=p_edge,noise_type=noise_type,strength_6qubit=strength_6qubit,noise_position=noise_position,samples=samples)
```

```python
# for creating a pdf
fig=tools.plot_pur_3reg_renyi(layers_max=layers_max,ns=ns,d=d,p_edge=p_edge,noise_type=noise_type,strength_6qubit=strength_6qubit,noise_position=noise_position,samples=samples)

plt.savefig("Fig_D1.pdf", format="pdf", bbox_inches="tight",dpi=300)
```
